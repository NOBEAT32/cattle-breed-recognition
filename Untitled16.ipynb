```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import os
import tqdm
import random
import cv2
from keras.preprocessing.image import load_img
warnings.filterwarnings('ignore')
import seaborn as sns
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator,img_to_array,load_img
from keras.utils import image_dataset_from_directory
```

```python

# Augmentation settings
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest'
)

target_images_per_class = 1500  # target count per breed

dataset_dir = "/content/Sih_prototype_data"

for class_name in os.listdir(dataset_dir):
    class_folder = os.path.join(dataset_dir, class_name)
    images = os.listdir(class_folder)
    count = len(images)

    print(f"Processing {class_name}... (Current: {count})")

    # Augment images until target is reached
    img_idx = 0
    while count < target_images_per_class:
        img_path = os.path.join(class_folder, images[img_idx % len(images)])  # cycle through existing images
        img = load_img(img_path)
        x = img_to_array(img)
        x = np.expand_dims(x, axis=0)

        # Generate one augmented image and save it in the SAME directory
        for batch in datagen.flow(x, batch_size=1,
                                  save_to_dir=class_folder,
                                  save_prefix=class_name,
                                  save_format="jpg"):
            count += 1
            break   # save only 1 new image per loop

        img_idx += 1

    print(f" {class_name} now has {count} images")
```

```python


input_path = []
label = []

dataset_dir = "/content/Sih_prototype_data"
class_names = os.listdir(dataset_dir)

class_to_idx = {name: idx for idx, name in enumerate(class_names)}

for class_name in class_names:
    class_folder = os.path.join(dataset_dir, class_name)
    for path in os.listdir(class_folder):
        input_path.append(os.path.join(class_folder, path))
        label.append(class_to_idx[class_name])

print("Total Images:", len(input_path))
print("First Image Path:", input_path[0])
print("First Label:", label[0])
print("Class Mapping:", class_to_idx)
```

```python
df = pd.DataFrame()
df['images'] = input_path
df['label'] = [os.path.basename(os.path.dirname(p)) for p in input_path]

df = df.sample(frac=1).reset_index(drop=True)
df.head()
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(df,test_size=0.2,random_state=42)
```

```python
train_generator = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    rotation_range=40,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
val_generator=ImageDataGenerator(rescale=1./255)

train_iterator = train_generator.flow_from_dataframe(
    train,
    x_col='images',
    y_col='label',
    target_size=(128,128),
    batch_size=32,
    class_mode='categorical'
)
test_iterator = train_generator.flow_from_dataframe(
    test,
    x_col='images',
    y_col='label',
    target_size=(128,128),
    batch_size=32,
    class_mode='categorical'
)
```

#Model creation


```python
from keras import Sequential
from keras.layers import Conv2D, MaxPool2D, Flatten, Dense

model = Sequential([
    Conv2D(16, (3,3), activation='relu', input_shape=(128,128,3)),
    MaxPool2D((2,2)),

    Conv2D(32, (3,3), activation='relu'),
    MaxPool2D((2,2)),

    Conv2D(64, (3,3), activation='relu'),
    MaxPool2D((2,2)),

    Flatten(),

    Dense(512, activation='relu'),
    Dense(8, activation='softmax')
])
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy']
)
model.summary()
```

```python
history=model.fit(train_iterator,epochs=10,validation_data=test_iterator)
```
